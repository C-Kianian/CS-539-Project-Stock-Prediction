```python
import yfinance as yf
import pandas as pd
import numpy as np
from ta.momentum import RSIIndicator, StochasticOscillator, WilliamsRIndicator
from ta.trend import MACD, SMAIndicator, CCIIndicator
from ta.volatility import BollingerBands, AverageTrueRange
from ta.volume import OnBalanceVolumeIndicator
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import matplotlib.pyplot as plt

tickers = ['AAPL', 'GOOG', 'BLK', 'NKE', 'NFLX', 'DIS']

results = []

for ticker in tickers:
    print(f"\nProcessing {ticker}...")
    df = yf.download(ticker, start='2010-01-01', end='2019-12-31')

    if df.empty or len(df) < 300:
        print(f"Not enough data for {ticker}. Skipping.")
        continue

    df.dropna(inplace=True)
    df['Return'] = df['Close'].pct_change()
    df['Volatility'] = df['Return'].rolling(window=5).std()

    close = df['Close'].squeeze()
    high = df['High'].squeeze()
    low = df['Low'].squeeze()
    volume = df['Volume'].squeeze()

    # === Basic Technical Indicators ===
    df['SMA_20'] = SMAIndicator(close, window=20).sma_indicator()
    df['SMA_50'] = SMAIndicator(close, window=50).sma_indicator()
    df['RSI'] = RSIIndicator(close, window=14).rsi()
    macd = MACD(close)
    df['MACD'] = macd.macd()
    df['MACD_diff'] = macd.macd_diff()
    df['MACD_signal'] = macd.macd_signal()
    bb = BollingerBands(close, window=20, window_dev=2)
    df['bb_bbm'] = bb.bollinger_mavg()
    df['bb_bbh'] = bb.bollinger_hband()
    df['bb_bbl'] = bb.bollinger_lband()
    stoch = StochasticOscillator(high, low, close, window=14, smooth_window=3)
    df['stoch_k'] = stoch.stoch()
    df['stoch_d'] = stoch.stoch_signal()
    obv = OnBalanceVolumeIndicator(close, volume)
    df['OBV'] = obv.on_balance_volume()

    # === Advanced Technical Indicators ===
    atr = AverageTrueRange(high, low, close, window=14)
    df['ATR'] = atr.average_true_range()
    cci = CCIIndicator(high, low, close, window=20)
    df['CCI'] = cci.cci()
    willr = WilliamsRIndicator(high, low, close, lbp=14)
    df['WilliamsR'] = willr.williams_r()

    # === Lagged returns ===
    for i in range(1, 4):
        df[f'return_lag_{i}'] = df['Return'].shift(i)

    # === Additional features ===
    df['volume_change'] = volume.pct_change()
    df['price_vol_ratio'] = df['Return'] / (volume + 1e-6)  # avoid div by zero

    # === Target: 1-day ahead price direction ===
    df['Target'] = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)

    df.dropna(inplace=True)

    # === Features to use ===
    feature_cols = ['SMA_20', 'SMA_50', 'RSI', 'MACD', 'MACD_diff', 'MACD_signal',
                    'bb_bbm', 'bb_bbh', 'bb_bbl', 'stoch_k', 'stoch_d', 'OBV',
                    'ATR', 'CCI', 'WilliamsR',
                    'return_lag_1', 'return_lag_2', 'return_lag_3',
                    'Volatility', 'volume_change', 'price_vol_ratio']

    X = df[feature_cols]
    y = df['Target']

    # === Feature scaling ===
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # === Train-test split ===
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, shuffle=False, test_size=0.2)

    # === Model ===
    clf = XGBClassifier(n_estimators=100000,
                        learning_rate=0.1,
                        max_depth=100000,
                        use_label_encoder=False,
                        eval_metric='logloss',
                        random_state=42)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    # === Metrics ===
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)

    print(f"Results for {ticker}:")
    print(classification_report(y_test, y_pred))

    results.append({
        'Ticker': ticker,
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'F1 Score': f1
    })

    # === Feature importance plot ===
    sorted_idx = np.argsort(clf.feature_importances_)[::-1]
    plt.figure(figsize=(10, 5))
    plt.title(f"{ticker} - XGBoost Feature Importance")
    plt.bar([feature_cols[i] for i in sorted_idx], clf.feature_importances_[sorted_idx])
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# === Summary ===
summary_df = pd.DataFrame(results)
summary_df.loc['Average'] = summary_df.mean(numeric_only=True)
print("\n=== Overall XGBoost Performance Summary ===")
print(summary_df)
```
